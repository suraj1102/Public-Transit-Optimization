```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("scarped_data/bus_stop_data.csv", index_col=0)
df = df.rename_axis("Sr No.").rename(columns={"Unnamed: 0": "Sr No.", "1": "Stop Name", "2": "Stop Code"})
df.index = pd.to_numeric(df.index, errors="coerce")
df.sort_index(inplace=True)
df
```

```python
df_filtered = df[~df["Stop Code"].str.contains("S", na=False)]
df_filtered.shape
```
